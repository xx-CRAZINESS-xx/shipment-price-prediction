```python
#mapping values of Frieght Cost based on ID

current_index=[]
replacement_freight_price_indx=[]
for indx,value in enumerate(df['Freight Cost (USD)']):
    if 'ID' in value:
        current_index.append(indx)
        string=value
        replacement_idx=re.findall('\d+',string)
        replacement_idx=replacement_idx[1]
        replacement_freight_price_indx.append(replacement_idx)
        
price=[]
for i in replacement_freight_price_indx:
    i=np.int(i)
    price.append(df[df['ID']==i]['Freight Cost (USD)'].values[0])
    
new_index=df['Freight Cost (USD)']
new_index[current_index]=price
```

```python
#mapping values of Weights based on ID

current_index=[]
replacement_weight_indx=[]
for indx,value in enumerate(df['Weight (Kilograms)']):
    if 'ID' in value:
        current_index.append(indx)
        string=value
        replacement_idx=re.findall('\d+',string)
        replacement_idx=replacement_idx[1]
        replacement_weight_indx.append(replacement_idx)
        
weights=[]
for i in replacement_weight_indx:
    i=np.int(i)
    weights.append(df[df['ID']==i]['Weight (Kilograms)'].values[0])
    
new_index=df['Weight (Kilograms)']
new_index[current_index]=weights
```

```python
# replacing string values as nan

df['Freight Cost (USD)']=df['Freight Cost (USD)'].replace('Freight Included in Commodity Cost',np.nan)
df['Freight Cost (USD)']=df['Freight Cost (USD)'].replace('Invoiced Separately',np.nan)

df['Weight (Kilograms)']=df['Weight (Kilograms)'].replace('Weight Captured Separately',np.nan)
```

```python
# imputing nan values based on same PQ #

df2=df.copy()
imputed=[]
for i in list(set(df2['PQ #'])):
    df_code=df2[df2['PQ #']==i]
    df_code['Weight (Kilograms)'].fillna(df_code['Weight (Kilograms)'].median(),inplace=True)
    df_code['Freight Cost (USD)'].fillna(df_code['Freight Cost (USD)'].median(),inplace=True)
    df_code['Line Item Insurance (USD)'].fillna(df_code['Line Item Insurance (USD)'].median(),inplace=True)
    df_code['Shipment Mode'].fillna(df_code['Shipment Mode'].mode(),inplace=True)
    imputed.append(df_code)
    df3=pd.concat(imputed)
```

```python
# imputing nan values based on same Project Code on df3

df4=df3.copy()
new_imputed=[]
for i in list(set(df4['Project Code'])):
    df_code=df4[df4['Project Code']==i]
    df_code['Weight (Kilograms)'].fillna(df_code['Weight (Kilograms)'].median(),inplace=True)
    df_code['Freight Cost (USD)'].fillna(df_code['Freight Cost (USD)'].median(),inplace=True)
    df_code['Line Item Insurance (USD)'].fillna(df_code['Line Item Insurance (USD)'].median(),inplace=True)   
    new_imputed.append(df_code)
    df5=pd.concat(new_imputed)
```

```python
# still some values has to be replaced

df5['Weight (Kilograms)'].fillna(df5['Weight (Kilograms)'].median(),inplace=True)
df5['Freight Cost (USD)'].fillna(df5['Freight Cost (USD)'].median(),inplace=True)
df5['Line Item Insurance (USD)'].fillna(df5['Line Item Insurance (USD)'].median(),inplace=True)

df5.isnull().sum()
```

```python
df5['Freight Cost (USD)']=df5['Freight Cost (USD)'].astype('float64')
df5['Weight (Kilograms)']=df5['Weight (Kilograms)'].astype('float64')
```
